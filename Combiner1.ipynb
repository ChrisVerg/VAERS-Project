# The purpose of this code should be to iterate the directory and convert all files to the uft-8 encoding

```python
import pandas as pd
import os
import glob
import chardet
```

```python
import codecs
```

```python
root_dir = 'C:/Users/chris/Desktop/WozU_Data_Science/VAERS-data'

for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        file_path = os.path.join(dirpath, filename)
        with codecs.open(file_path, 'r', encoding='ISO-8859-1') as f:
            content = f.read()
        with codecs.open(file_path, 'w', encoding='UTF-8') as f:
            f.write(content)
```

### The 1998 data previously would not read in as CSV due to encoding. The above code converted all data files to encoding UTF-8 so they can all be merged.
